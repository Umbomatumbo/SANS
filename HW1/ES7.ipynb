```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, geom, poisson, uniform, expon, norm, beta

# Set the number of samples
M = 10000

# Initialize random number generator for reproducibility
np.random.seed(0)

# Helper function to calculate and print estimated mean and variance
def calculate_stats(samples, title):
    estimated_mean = np.mean(samples)
    estimated_variance = np.var(samples)
    print(f"{title}:")
    print(f"Estimated Mean: {estimated_mean:.4f}")
    print(f"Estimated Variance: {estimated_variance:.4f}")
    #print the theoretical mean and variance
    if title == "Bernoulli":
        print(f"Theoretical Mean: {p_a:.4f}")
        print(f"Theoretical Variance: {p_a*(1-p_a):.4f}")
    elif title == "Binomial":
        print(f"Theoretical Mean: {n_b*p_b:.4f}")
        print(f"Theoretical Variance: {n_b*p_b*(1-p_b):.4f}")
    elif title == "Geometric":
        print(f"Theoretical Mean: {1/p_c:.4f}")
        print(f"Theoretical Variance: {(1-p_c)/(p_c**2):.4f}")
    elif title == "Poisson":
        print(f"Theoretical Mean: {lambda_d:.4f}")
        print(f"Theoretical Variance: {lambda_d:.4f}")
    elif title == "Uniform":
        print(f"Theoretical Mean: {0.5:.4f}")
        print(f"Theoretical Variance: {1/12:.4f}")
    elif title == "Exponential":
        print(f"Theoretical Mean: {1/lambda_f:.4f}")
        print(f"Theoretical Variance: {1/lambda_f**2:.4f}")
    elif title == "Gaussian (µ=1, σ^2=1)":
        print(f"Theoretical Mean: {mu_g:.4f}")
        print(f"Theoretical Variance: {sigma_g**2:.4f}")
    elif title == "Gaussian (µ=1, σ^2=5)":
        print(f"Theoretical Mean: {mu_h:.4f}")
        print(f"Theoretical Variance: {sigma_h**2:.4f}")
    elif title == "Beta (α=1, β=5)":
        print(f"Theoretical Mean: {alpha_i/(alpha_i+beta_i):.4f}")
        print(f"Theoretical Variance: {(alpha_i*beta_i)/((alpha_i+beta_i)**2*(alpha_i+beta_i+1)):.4f}")

    return estimated_mean, estimated_variance

# a) Bernoulli with p = 3/4
p_a = 3/4
samples_a = bernoulli.rvs(p_a, size=M)
plt.figure(figsize=(15, 15))
plt.subplot(3, 3, 1)
plt.plot(range(1, 101), samples_a[:100])
plt.title("Bernoulli Samples")
plt.subplot(3, 3, 2)
plt.hist(samples_a, bins=20, density=True, cumulative=True)
plt.title("CDF")
plt.subplot(3, 3, 3)
plt.hist(samples_a, bins=2, density=True)
plt.title("PMF")
mean_a, var_a = calculate_stats(samples_a, "Bernoulli")
#estimate expected value and variance
mean_a = np.mean(samples_a)
var_a = np.var(samples_a)



# b) Binomial with n=10 and p=3/4
n_b = 10
p_b = 3/4
samples_b = binom.rvs(n_b, p_b, size=M)
plt.subplot(3, 3, 4)
plt.plot(range(1, 101), samples_b[:100])
plt.title("Binomial Samples")
plt.subplot(3, 3, 5)
plt.hist(samples_b, bins=20, density=True, cumulative=True)
plt.title("CDF")
plt.subplot(3, 3, 6)
plt.hist(samples_b, bins=30, density=True)
plt.title("PMF")
mean_b, var_b = calculate_stats(samples_b, "Binomial")

# c) Geometric with p=3/4
p_c = 3/4
samples_c = geom.rvs(p_c, size=M)
plt.subplot(3, 3, 7)
plt.plot(range(1, 101), samples_c[:100])
plt.title("Geometric Samples")
plt.subplot(3, 3, 8)
plt.hist(samples_c, bins=20, density=True, cumulative=True)
plt.title("CDF")
plt.subplot(3, 3, 9)
plt.hist(samples_c, bins=15, density=True)
plt.title("PMF")
mean_c, var_c = calculate_stats(samples_c, "Geometric")

# d) Poisson with λ = 1
lambda_d = 1
samples_d = poisson.rvs(lambda_d, size=M)
plt.figure(figsize=(15, 15))
plt.subplot(3, 3, 1)
plt.plot(range(1, 101), samples_d[:100])
plt.title("Poisson Samples")
plt.subplot(3, 3, 2)
plt.hist(samples_d, bins=20, density=True, cumulative=True)
plt.title("CDF")
plt.subplot(3, 3, 3)
plt.hist(samples_d, bins=15, density=True)
plt.title("PMF")
mean_d, var_d = calculate_stats(samples_d, "Poisson")

# e) Uniform in [0, 1]
samples_e = uniform.rvs(size=M)
plt.subplot(3, 3, 4)
plt.plot(range(1, 101), samples_e[:100])
plt.title("Uniform Samples")
plt.subplot(3, 3, 5)
plt.hist(samples_e, bins=20, density=True, cumulative=True)
plt.title("CDF")
plt.subplot(3, 3, 6)
plt.hist(samples_e, bins=20, density=True)
plt.title("PDF")
mean_e, var_e = calculate_stats(samples_e, "Uniform")

# f) Exponential with λ = 1
lambda_f = 1
samples_f = expon.rvs(scale=1/lambda_f, size=M)
plt.subplot(3, 3, 7)
plt.plot(range(1, 101), samples_f[:100])
plt.title("Exponential Samples")
plt.subplot(3, 3, 8)
plt.hist(samples_f, bins=20, density=True, cumulative=True)
plt.title("CDF")
plt.subplot(3, 3, 9)
plt.hist(samples_f, bins=20, density=True)
plt.title("PDF")
mean_f, var_f = calculate_stats(samples_f, "Exponential")

# g) Univariate Gaussian with µ = 1, σ^2 = 1
mu_g = 1
sigma_g = 1
samples_g = norm.rvs(loc=mu_g, scale=sigma_g, size=M)
plt.figure(figsize=(15, 15))
plt.subplot(3, 3, 1)
plt.plot(range(1, 101), samples_g[:100])
plt.title("Gaussian Samples (µ=1, σ^2=1)")
plt.subplot(3, 3, 2)
plt.hist(samples_g, bins=20, density=True, cumulative=True)
plt.title("CDF")
plt.subplot(3, 3, 3)
plt.hist(samples_g, bins=30, density=True)
plt.title("PDF")
mean_g, var_g = calculate_stats(samples_g, "Gaussian (µ=1, σ^2=1)")

# h) Univariate Gaussian with µ = 1, σ^2 = 5
mu_h = 1
sigma_h = np.sqrt(5)
samples_h = norm.rvs(loc=mu_h, scale=sigma_h, size=M)
plt.subplot(3, 3, 4)
plt.plot(range(1, 101), samples_h[:100])
plt.title("Gaussian Samples (µ=1, σ^2=5)")
plt.subplot(3, 3, 5)
plt.hist(samples_h, bins=20, density=True, cumulative=True)
plt.title("CDF")
plt.subplot(3, 3, 6)
plt.hist(samples_h, bins=30, density=True)
plt.title("PDF")
mean_h, var_h = calculate_stats(samples_h, "Gaussian (µ=1, σ^2=5)")

# i) Beta with α = 1, β = 5
alpha_i = 1
beta_i = 5
samples_i = beta.rvs(alpha_i, beta_i, size=M)
plt.subplot(3, 3, 7)
plt.plot(range(1, 101), samples_i[:100])
plt.title("Beta Samples (α=1, β=5)")
plt.subplot(3, 3, 8)
plt.hist(samples_i, bins=20, density=True, cumulative=True)
plt.title("CDF")
plt.subplot(3, 3, 9)
plt.hist(samples_i, bins=30, density=True)
plt.title("PDF")
mean_i, var_i = calculate_stats(samples_i, "Beta (α=1, β=5)")

# j) Sum of independent samples from f) and h)
samples_f_h = samples_f + samples_h
plt.figure()
plt.subplot(1, 2, 1)
plt.plot(range(1, 101), samples_f_h[:100])
plt.title("Sum of Gaussian Samples (f) and h)")
plt.subplot(1, 2, 2)
plt.hist(samples_f_h, bins=40, density=True, cumulative=True)
plt.title("CDF")

# Show plots
plt.tight_layout()
plt.show()
```
