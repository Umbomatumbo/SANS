```python
#import numpy
import numpy as np
#create a matrix matri AN with vales
# 4 3 2 1
# 1 2 1 3
# 3 1 0 2
# 2 3 4 5
AN = np.matrix([[4,3,2,1],[1,2,1,3],[3,1,0,2],[2,3,4,5]])
#compute the eigenvalues of matrix AN
eigvals, eigvecs = np.linalg.eig(AN)
#create a matrix matri Van with aigenvales as coloms
Van = np.matrix(eigvecs)
#chek if the Van is orthogonal
if np.allclose(Van.T, Van.I):
    print("Van is orthogonal")
else:
    print("Van is not orthogonal")
#define Lan as the diagonal matrix of eigenvalues of AN
Lan = np.matrix(np.diag(eigvals))
#chek if the Lan is orthogonal
if np.allclose(Lan.T, Lan.I):
    print("Lan is orthogonal")
else:
    print("Lan is not orthogonal")

#create a matrix Bn with vales
# 5 0 2 −1
# 0 2 0 −1
# 2 0 1 0
# −1 −1 0 2 
Bn = np.matrix([[5,0,2,-1],[0,2,0,-1],[2,0,1,0],[-1,-1,0,2]])
#compute the eigenvalues of matrix Bn
eigvals, eigvecs = np.linalg.eig(Bn)
#create a matrix matri Vbn with aigenvales as coloms
Vbn = np.matrix(eigvecs)
#chek if the Vbn is orthogonal
if np.allclose(Vbn.T, Vbn.I):
    print("Vbn is orthogonal")
else:
    print("Vbn is not orthogonal")
#define Lbn as the diagonal matrix of eigenvalues of Bn
Lbn = np.matrix(np.diag(eigvals))
#chek if the Lbn is orthogonal
if np.allclose(Lbn.T, Lbn.I):
    print("Lbn is orthogonal")
else:
    print("Lbn is not orthogonal")  

#create a matrix Cn with vales
# 2 −1 0
# −1 2 −1
# 0 −1 2
Cn= np.matrix([[2,-1,0],
               [-1,2,-1],
               [0,-1,2]])
#compute the eigenvalues of matrix Cn
eigvals, eigvecs = np.linalg.eig(Cn)
#create a matrix matri Vcn with aigenvales as coloms
Vcn = np.matrix(eigvecs)
#chek if the Vcn is orthogonal
if np.allclose(Vcn.T, Vcn.I):
    print("Vcn is orthogonal")
else:
    print("Vcn is not orthogonal")
#define Lcn as the diagonal matrix of eigenvalues of Cn
Lcn = np.matrix(np.diag(eigvals))
#chek if the Lcn is orthogonal
if np.allclose(Lcn.T, Lcn.I):
    print("Lcn is orthogonal")
else:
    print("Lcn is not orthogonal")



#create a matrix Dn with vales
# 4 3
# 3 1
Dn= np.matrix([[4,3],[3,1]])
#compute the eigenvalues of matrix Dn
eigvals, eigvecs = np.linalg.eig(Dn)
#create a matrix matri Vdn with aigenvales as coloms
Vdn = np.matrix(eigvecs)
#chek if the Vdn is orthogonal
if np.allclose(Vdn.T, Vdn.I):
    print("Vdn is orthogonal")
else:
    print("Vdn is not orthogonal")
#define Ldn as the diagonal matrix of eigenvalues of Dn
Ldn = np.matrix(np.diag(eigvals))

#chek if the Ldn is orthogonal
if np.allclose(Ldn.T, Ldn.I):
    print("Ldn is orthogonal")
else:
    print("Ldn is not orthogonal")

```

```python
#defien a metrix E

# 4 3 2 1
# 1 2 1 3
# 3 1 0 2
# 2 3 4 5
# 2 1 3 2

E = Matrix([[4,3,2,1],[1,2,1,3],[3,1,0,2],[2,3,4,5],[2,1,3,2]])
#define the projecton matrix P
Q= E.T*E
P= E*Q.inv()*E.T
#define the vector b st b.t= 1,2,1,5,5
b = Matrix([[1],[2],[1],[5],[5]])
#compute the projection of b on E
print("The projection of b on E is:")
pprint(P*b)

#defien a metrix F
 
 #5 0 2 −1
 #0 2 0 −1
 #2 0 1 0
 #−1 −1 0 2

F = Matrix([[5,0,2,-1],[0,2,0,-1],[2,0,1,0],[-1,-1,0,2]])
#define the projecton matrix P
Q= F.T*F
P= F*Q.inv()*F.T
#define the vector b st b.t= 1,2,1,5
b = Matrix([[1],[2],[1],[5]])
#compute the projection of b on F
print("The projection of b on F is:")
pprint(P*b)



```

```python
import matplotlib.pyplot as plt
import numpy as np

# Definisci i vettori
vectors = np.array([[1, 0], [0, 1], [-1, 0], [0, -1]])

# Calcola i punti sulla circonferenza unitaria
angles = np.linspace(0, 2 * np.pi, 100)  # Usiamo 100 punti per ottenere una circonferenza più liscia
x = np.cos(angles)
y = np.sin(angles)

# Crea un grafico
plt.figure(figsize=(6, 6))

# Traccia la circonferenza unitaria
plt.plot(x, y, 'b-', label='Unit circle')

# Aggiungi i vettori
for i in range(4):
    plt.quiver(0, 0, vectors[i, 0], vectors[i, 1], angles='xy', scale_units='xy', scale=1, color=['g', 'r', 'c', 'm'][i], label=f'Vector {i+1}')

# Imposta l'aspetto del grafico
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
#print the legend in the top left corner
plt.legend(loc='upper left')

# Mostra il grafico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the matrix A
A = np.array([[1, 0], [-1, 2]])

# Generate points on the unit circle
theta = np.linspace(0, 2*np.pi, 100)
unit_circle = np.array([np.cos(theta), np.sin(theta)])

# Apply the matrix transformation to the unit circle points
transformed_points = np.dot(A, unit_circle)

# Plot the unit circle
plt.plot(unit_circle[0], unit_circle[1], label='Unit Circle')

#calculate the eigenvalues and eigenvectors of A
eigvals, eigvecs = np.linalg.eig(A)

# scale eigenvectors to match the transformed unit circle
eigvecs_scaled = eigvecs * np.sqrt(eigvals)

#print the eigenvalues and eigenvectors
print("The eigenvalues of A are:")
print(eigvals)
print("The eigenvectors of A are:")
print(eigvecs)
print("The scaled eigenvectors of A are:")
print(eigvecs_scaled)

# Plot the transformed ellipse
plt.plot(transformed_points[0], transformed_points[1], label='Transformed Ellipse')


plt.plot(1,-1 , 'bo', label="Point (1,-1)")
plt.plot(-1,1, 'bo', label="Point (-1,1)")

plt.plot(0,2, 'bo', label="Point (0,2)")
plt.plot(0,-2, 'bo', label="Point (0,-2)")

plt.plot(eigvecs_scaled[1, 0],eigvecs_scaled[1, 1] , 'bo', label="Point ("+str(eigvecs_scaled[1, 0])+","+str(eigvecs_scaled[1, 0])+")")
# Set equal scaling to make sure the circle and ellipse are visually accurate
plt.axis('equal')

# Set labels and legend
plt.title('Unit Circle and Transformed Ellipse')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
#show grid
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Show the plot
plt.show()
```
