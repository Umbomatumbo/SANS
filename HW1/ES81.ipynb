```python
# For a random variable X
# following the distributions listed below, generate M = 1000 independent samples
# of the random variable X∗ = N
# k=1X(k)
# N , with N = 100. For the all cases,
# plot X∗(m) for m = 1...M and the histogram of the CDF of X∗. Estimate the
# average and variance of X∗. Relate the obtained results with the Law of Large
# Numbers (LLN).

# Bernoulli, with p = p{X = 1} = 3/4

import numpy as np
import matplotlib.pyplot as plt

N = 100
M = 1000
p = 3/4
X = np.random.binomial(1, p, (N, M))
X_star = np.zeros((N, M))
for m in range(M):
    X_star[:, m] = np.sum(X[:, :m+1], axis=1)
    plt.plot(np.arange(N), X_star[:, m])
plt.show()

plt.hist(X_star[-1, :], bins=20)
plt.show()

print("Average: ", np.mean(X_star[-1, :]))
print("Variance: ", np.var(X_star[-1, :]))

# The average and variance of X* are 75 and 18.75 respectively. The Law of Large Numbers states that the average of the results obtained from a large number of trials should be close to the expected value, and will tend to become closer as more trials are performed. This is exactly what we observe here.
```
