```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
p = 3/4  # Probability of success for the Bernoulli random variable
N = 100  # Number of samples
M = 1000  # Number of experiments

# Simulate M experiments of N Bernoulli trials
experiments = np.random.binomial(1, p, size=(M, N))

# Calculate the sum of N trials for each experiment
X_star = np.sum(experiments, axis=1)

# Plot X*(m) for m = 1...M
plt.figure(figsize=(10, 4))
plt.plot(range(1, M + 1), X_star, marker='o', linestyle='-', color='b')
plt.xlabel('Experiment Number (m)')
plt.ylabel('X*(m)')
plt.title('X*(m) for Bernoulli Random Variable')
plt.grid(True)

# Calculate the histogram of the CDF of X*
plt.figure(figsize=(10, 4))
plt.hist(X_star, bins=50, density=True, cumulative=True, color='g', alpha=0.7)
plt.xlabel('X*')
plt.ylabel('CDF of X*')
plt.title('Histogram of the CDF of X*')
plt.grid(True)

# Calculate the average and variance of X*
average_X_star = np.mean(X_star/100)
variance_X_star = np.var(X_star/100)

print("Average of X*: {:.2f}".format(average_X_star))
print("Variance of X*: {:.2f}".format(variance_X_star))

# Display the plots
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu = 1  # Mean of the Gaussian distribution
sigma = 1  # Standard deviation (square root of variance)
N = 100  # Number of samples
M = 1000  # Number of experiments

# Simulate M experiments of N Gaussian random variables with mean mu and variance sigma^2
experiments = np.random.normal(mu, sigma, size=(M, N))

# Calculate the sum of N values for each experiment
X_star = np.sum(experiments, axis=1)

# Plot X*(m) for m = 1...M
plt.figure(figsize=(10, 4))
plt.plot(range(1, M + 1), X_star, marker='o', linestyle='-', color='b')
plt.xlabel('Experiment Number (m)')
plt.ylabel('X*(m)')
plt.title('X*(m) for Gaussian Random Variable (=1, ^2=1)')
plt.grid(True)

# Calculate the histogram of the CDF of X*
plt.figure(figsize=(10, 4))
plt.hist(X_star, bins=20, density=True, cumulative=True, color='g', alpha=0.7)
plt.xlabel('X*')
plt.ylabel('CDF of X*')
plt.title('Histogram of the CDF of X*')
plt.grid(True)

# Calculate the average and variance of X*
average_X_star = np.mean(X_star)
variance_X_star = np.var(X_star)

print("Average of X*: {:.2f}".format(average_X_star))
print("Variance of X*: {:.2f}".format(variance_X_star))

# Display the plots
plt.show()
```
