```python

# import sympy  
from sympy import * 
```

```python
# 1. Define a matrix A with size (2,4) with value
 #4 3 2 1
 #3 1 4 2

A = Matrix([[4,3,2,1],
            [3,1,4,2]]) # 2x4 matrix
pprint(A)

# 2. Define a matrix B
# 4 3 2
# 3 1 5
# 1 4 7
# 5 2 6
# 3 4 2

B = Matrix([[4,3,2],
            [3,1,5],
            [1,4,7],
            [5,2,6],
            [3,4,2]]) # 5x3 matrix
pprint(B)
```

```python
# 3. Calculate the following A
A_rowspace = A.rowspace()
#calculate the row space of A
A_columnspace = A.columnspace()
#calculate the null space of A
A_nullspace = A.nullspace()
# calculate the left null space of A
A_leftnullspace = (A.T).nullspace()
#calculate the rank of A
A_rank = A.rank()

# calculate the row space of B
B_rowspace = B.rowspace()
# calculate the column space of B
B_columnspace = B.columnspace()
# calculate the null space of B
B_nullspace = B.nullspace()
#c  calcuate the left null space of B
B_leftnullspace = (B.T).nullspace()


# print the results
pprint("The column space of A is: ")
pprint(A_columnspace)
pprint("The row space of A is: ")
pprint(A_rowspace)
pprint("The null space of A is: ")
pprint(A_nullspace)
pprint("The left null space of A is: ")
pprint(A_leftnullspace)
pprint("The rank of A is: ")
pprint(A_rank)



#print same empty line
print()



print("The row space of B is: ")
pprint(B_rowspace)
print("The column space of B is: ")
pprint(B_columnspace)
print("The null space of B is: ")
pprint(B_nullspace)
print("The left null space of B is: ")
pprint(B_leftnullspace)

```

```python
# Caluate the right pseudoinverse of A
A_pseudo_inverse = A.pinv()
print("The right pseudoinverse of A is: ")
pprint(A_pseudo_inverse)
# Caluate the left pseudoinverse of B
B_pseudo_inverse = B.pinv()
print("The left pseudoinverse of B is: ")
pprint(B_pseudo_inverse)
```

```python
#calculate the moltiplication of pseudoinverse of A and  vector b =
# 2
# 1

b = Matrix([[2],[1]])
# molitiply the pseudoinverse of A and b
x=A_pseudo_inverse*b
print("The moltiplication of pseudoinverse of A and vector b is: ")
pprint(x)
#calculate the moltiplication of pseudoinverse of B and  vector c =
# 1
# 2
# 1
# 5
# 4

c = Matrix([[1],[2],[1],[5],[4]])
# molitiply the pseudoinverse of B and c
y=B_pseudo_inverse*c
print("The moltiplication of pseudoinverse of B and vector c is: ")
pprint(y)
```
